```python
import pandas as pd
ny = pd.read_csv('https://raw.githubusercontent.com/Datamanim/pandas/main/AB_NYC_2019.csv')
```

## 44번
데이터를 로드하고 상위 5개 컬럼을 출력하라

```python
ny.head()
```

## 45번
데이터의 각 host_name의 빈도수를 구하고 host_name으로 정렬하여 상위 5개를 출력하라

```python
ans = ny.host_name.value_counts().sort_index()
ans.head()
```

## 46번
 데이터의 각 host_name의 빈도수를 구하고 빈도수 기준 내림차순 정렬한 데이터 프레임을 만들어라. 빈도수 컬럼은 counts로 명명하라

```python
ans = ny.groupby('host_name').size().to_frame().rename(columns={0:'counts'}).sort_values('counts', ascending=False)
ans.head()
```

## 47번
neighbourhood_group의 값에 따른 neighbourhood컬럼 값의 갯수를 구하여라

```python
Ans = ny.groupby(['neighbourhood_group','neighbourhood'], as_index=False).size()
Ans
```

## 48번
neighbourhood_group의 값에 따른 neighbourhood컬럼 값 중 neighbourhood_group그룹의 최댓값들을 출력하라

```python
Ans = ny.groupby(['neighbourhood_group','neighbourhood'], as_index=False).size().groupby(['neighbourhood_group'], as_index=False).max()
Ans
```

## 49번
neighbourhood_group 값에 따른 price값의 평균, 분산, 최대, 최소 값을 구하여라

```python
#방법 1
ans = ny[['neighbourhood_group','price']].groupby('neighbourhood_group')
ans.describe()
```

```python
#방법 2
ans1 = ny[['neighbourhood_group','price']].groupby('neighbourhood_group').agg(['mean','var','max','min'])
ans1
```

## 50번
neighbourhood_group 값에 따른 reviews_per_month 평균, 분산, 최대, 최소 값을 구하여라

```python
ans = ny[['neighbourhood_group','reviews_per_month']].groupby('neighbourhood_group').agg(['mean','var','max','min'])
ans
```

## 51번
neighbourhood 값과 neighbourhood_group 값에 따른 price 의 평균을 구하라

```python
ans = ny.groupby(['neighbourhood_group','neighbourhood']).price.mean()
ans
```

## 52번
neighbourhood 값과 neighbourhood_group 값에 따른 price 의 평균을 계층적 indexing 없이 구하라

```python
ans = ny.groupby(['neighbourhood_group','neighbourhood']).price.mean().unstack()
ans
```

## 53번
neighbourhood 값과 neighbourhood_group 값에 따른 price 의 평균을 계층적 indexing 없이 구하고 nan 값은 -999값으로 채워라

```python
ans = ny.groupby(['neighbourhood', 'neighbourhood_group']).price.mean().unstack()
ans = ans.fillna(-999)
ans
```

## 54번
데이터중 neighbourhood_group 값이 Queens값을 가지는 데이터들 중 neighbourhood 그룹별로 price값의 평균, 분산, 최대, 최소값을 구하라

```python
ans = ny[ny.neighbourhood_group=='Queens'].groupby(['neighbourhood']).price.agg(['mean','var','max','min'])
ans
```

## 55번
데이터중 neighbourhood_group 값에 따른 room_type 컬럼의 숫자를 구하고 neighbourhood_group 값을 기준으로 각 값의 비율을 구하여라

```python
Ans = ny[['neighbourhood_group','room_type']].groupby(['neighbourhood_group','room_type']).size().unstack()
Ans.loc[:,:] = (Ans.values /Ans.sum(axis=1).values.reshape(-1,1))
```
