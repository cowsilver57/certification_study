```python
import pandas as pd
data = pd.read_csv('https://raw.githubusercontent.com/Datamanim/pandas/main/timeTest.csv')
```

#64번
데이터를 로드하고 각 열의 데이터 타입을 파악하라

```python
data.info()
```

# 65번
Yr_Mo_Dy을 판다스에서 인식할 수 있는 datetime64타입으로 변경하라

```python
data['Yr_Mo_Dy'] = pd.to_datetime(data['Yr_Mo_Dy'])
ans = data['Yr_Mo_Dy']
print(ans)
```

# 66번
Yr_Mo_Dy에 존재하는 년도의 유일값을 모두 출력하라

```python
ans = data['Yr_Mo_Dy'].dt.year.unique()
print(ans)
```

# 67번
Yr_Mo_Dy에 년도가 2061년 이상의 경우에는 모두 잘못된 데이터이다. 해당경우의 값은 100을 빼서 새롭게 날짜를 Yr_Mo_Dy 컬럼에 정의하라


```python
def fix_century(x):
    import datetime

    year = x.year - 100 if x.year >= 2061 else x.year
    return pd.to_datetime(datetime.date(year, x.month, x.day))

data['Yr_Mo_Dy'] = data['Yr_Mo_Dy'].apply(fix_century)
data.head()
```

# 68번
년도별 각컬럼의 평균값을 구하여라

```python
ans = data.groupby(data.Yr_Mo_Dy.dt.year).mean()
ans.head()
```

# 69번
weekday컬럼을 만들고 요일별로 매핑하라 ( 월요일: 0 ~ 일요일 :6)

```python
data['weekday'] = data.Yr_Mo_Dy.dt.weekday
data['weekday'].head()
```

# 70번
weekday컬럼을 기준으로 주말이면 1 평일이면 0의 값을 가지는 WeekCheck 컬럼을 만들어라

```python
data['WeekCheck'] = data['weekday'].map(lambda x : 1 if x in [5,6] else 0)
data['WeekCheck'].head()
```

# 71번
년도, 일자 상관없이 모든 컬럼의 각 달의 평균을 구하여라

```python
ans = data.groupby(data.Yr_Mo_Dy.dt.month).mean()
ans.head()
```

# 72번
모든 결측치는 컬럼기준 직전의 값으로 대체하고 첫번째 행에 결측치가 있을경우 뒤에있는 값으로 대채하라

```python
data = data.fillna('ffill').fillna('bfill')
#ffill = 앞으로 채우기
#bfill = 뒤로 채우기
data.isnull().sum()
```

# 73번
년도 - 월을 기준으로 모든 컬럼의 평균값을 구하여라

```python
Ans = data.groupby(data.Yr_Mo_Dy.dt.to_period('M')).mean()
#to_period = 날짜를 월 단위 기간으로 변환
ans.head()
```

# 74번
RPT 컬럼의 값을 일자별 기준으로 1차차분하라

```python
# info를 활용하여 RPT 값이 명목형 변수임을 확인하고 인코딩 진행
# 'RPT' 열에 있는 문자열 값을 숫자로 변환
data['RPT'] = pd.to_numeric(data['RPT'], errors='coerce')

# 변환된 'RPT' 열을 사용하여 차분(diff)을 계산
ans = data['RPT'].diff()
ans
```

# 75번
RPT와 VAL의 컬럼을 일주일 간격으로 각각 이동평균한값을 구하여라

```python
ans = data[['RPT','VAL']].rolling(7).mean()
#rolling = 이동평균을 계산하는 메서드
ans.head(10)
```
