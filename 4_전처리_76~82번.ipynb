```python
import pandas as pd
data = pd.read_csv('https://raw.githubusercontent.com/Datamanim/pandas/main/seoul_pm.csv')
```

# 76번
년-월-일:시 컬럼을 pandas에서 인식할 수 있는 datetime 형태로 변경하라. (서울시의 제공데이터의 경우 0시가 24시로 표현된다)

```python
def change_date(x):
    import datetime
    # ':' 기준으로 분리한 문자열에서 시간 부분을 추출
    hour = x.split(':')[1]
    # ':' 기준으로 분리한 문자열에서 날짜 부분을 추출
    date = x.split(":")[0]

    # 만약 시간이 '24'라면 시간을 '00:00:00'으로 변경
    if hour =='24':
        hour ='00:00:00'

        # 날짜와 시간을 조합하여 datetime 객체를 생성하고, 하루를 더해줌.
        FinalDate = pd.to_datetime(date +" "+hour) + datetime.timedelta(days=1)

    # 시간이 '24'가 아니라면 시간을 'HH:00:00' 형식으로 변경
    else:
        hour = hour +':00:00'

        # 날짜와 시간을 조합하여 datetime 객체를 생성
        FinalDate = pd.to_datetime(date +" "+hour)
    # 최종 변환된 datetime 객체를 반환
    return FinalDate
# 'change_date' 함수를 각 행에 적용하여 변환된 값을 열에 적용
data['(년-월-일:시)'] = data['(년-월-일:시)'].apply(change_date)

ans = data
ans.head()
```

# 77번

```python
data['dayName'] = data['(년-월-일:시)'].dt.day_name()
data['dayName'].head()
```

# 78번

```python
ans = data.groupby('dayName')['PM10등급'].value_counts()
ans
```
