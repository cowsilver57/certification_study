# 83~86번

```python
import pandas as pd
data1 = pd.read_csv('https://raw.githubusercontent.com/Datamanim/pandas/main/under5MortalityRate.csv')
```

## 83번
Indicator을 삭제하고 First Tooltip 컬럼에서 신뢰구간에 해당하는 표현을 지워라

```python
data1.head()
```

```python
data1.drop('Indicator', axis=1, inplace=True)
data1['First Tooltip'] = data1['First Tooltip'].str.split("[").str[0].astype(float)
data1.head()
```

## 84번
년도가 2015년 이상, Dim1이 Both sexes인 케이스만 추출하라

```python
ans84 = data1[(data1.Period >= 2015)&(data1.Dim1 == 'Both sexes')]
ans84.head(3)
```

## 85번
84번 문제에서 추출한 데이터로 아래와 같이 나라에 따른 년도별 사망률을 데이터 프레임화 하라

```python
ans = ans84.pivot(index='Location', columns='Period', values='First Tooltip')
ans.head(3)
```

## 86번
Dim1에 따른 년도별 사망비율의 평균을 구하라

```python
ans = data1.pivot_table(index='Dim1', columns='Period', values='First Tooltip', aggfunc='mean')
ans.head(3)
```

# 87~90번

```python
data2 = pd.read_csv('https://raw.githubusercontent.com/Datamanim/pandas/main/winter.csv')
data2.head(3)
```

## 87번
데이터에서 한국 KOR 데이터만 추출하라

```python
ans87 = data2[data2.Country == 'KOR']
ans87.head(3)
```

## 88번
한국 올림픽 메달리스트 데이터에서 년도에 따른 medal 갯수를 데이터프레임화 하라

```python
ans = ans87.pivot_table(index='Year', columns='Medal', aggfunc='size')
ans.head(3)
```

## 89번
전체 데이터에서 sport종류에 따른 성별수를 구하여라

```python
ans = data2.pivot_table(index='Sport', columns='Gender', aggfunc='size')
ans.head()
```

## 90번
전체 데이터에서 Discipline종류에 따른 따른 Medal수를 구하여라

```python
ans = data2.pivot_table(index='Discipline', columns='Medal', aggfunc='size')
ans.head(383)
```
